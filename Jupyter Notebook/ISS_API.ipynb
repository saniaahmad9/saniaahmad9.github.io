```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
```

```python
url = "http://api.open-notify.org/iss-now.json"
query_url = f"{url}"
print(query_url)
response = requests.get(query_url).json()
```

```python
location = response["iss_position"]
lat = location["latitude"]
lon = location["longitude"]
timestamp = response["timestamp"]
    
#Output informationon screen
print("Latitude: " + str(lat))
print("Longitude: " +str(lon))
print("Time: " +str(timestamp))

#refresh position every 5 seconds
time.sleep(25)
```

```python
lat1 = []
long1 = []
timestamp1 = []
counter = 1

while counter > 0 :
    response = requests.get(query_url).json()
    lat1.append(response["iss_position"]["latitude"])
    long1.append(response["iss_position"]["longitude"])
    timestamp1.append(response["timestamp"])
    counter+=1
    time.sleep(30)
```
